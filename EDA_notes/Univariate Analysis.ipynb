### Now let us have a closer look at the distribution by plotting a simple histogram with 10 bins.

```python
plt.figure(figsize=(20,10)) # makes the plot wider
plt.hist(data, color='g') # plots a simple histogram
plt.axvline(data.mean(), color='m', linewidth=1)
plt.axvline(data.median(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(data.mode()[0], color='w', linestyle='dashed', linewidth=1)
```

- In the above histogram we can see that the second bin with class limits (4.81,  9.62) has the most number of observations. 
- A bulk of the observations lie within the first four classes. The rest of the six classes contain only a very few observations.
- In this case we are measuring distances. We can safely say based on visual observation that most of the distance present on the data are less than 20 kms.
- There are a very few distances that are more than 20 kms. Once we pass the 30 km point the number of observations dlrops further.
- From the above figure we can see the the mean is represented by the magenta line and the mode by the white line . The median is represented by the blue line.
- We can see from the above figure that the mode and the mean are very close to each other and that the median is lower than both.

#### <span style="font-family: Arial; font-weight:bold;font-size:1.9em;color:#0e92ea"> Handling non-numeric data

**Most of the machine learning models are designed to work on numeric data. Hence, we need to convert categorical text data into numerical data for model building.**

#### <span style="font-family: Arial; font-weight:bold;font-size:1.9em;color:#0e92ea">One Hot Encoding

**One-Hot-Encoding is used to create dummry variables to replace the categories in a categorical variable into features of each category and represent it using 1 or 0 based on the presence or absence of the categorical value in the record.**

### Pandas Get_Dummies

**This function does One-Hot-Encoding on categorical text**

```python
df_dummies= pd.get_dummies(data1, prefix='Park', columns=['ParkingArea'])
```

```python
df_dummies.head(5)
```

```python
df_dummies.columns
```

### Convert to number

**Sometimes if there are too many categories the, it is better to go for label encoding rather than using dummy variables.**

### sklearn LabelEncoder

**This function converts categorical text into categorical numbers.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
df_dummies['RegionId'] = labelencoder.fit_transform(df_dummies.Regionname)
```

```python
df_dummies['RegionId'].head(5)
```

```python
df_dummies['RegionId'].value_counts()
```

```python
df_dummies['RegionId'].unique()
```

```python
df_dummies['RegionId'].nunique()
```

### sklearn OneHotEncoder

**This function does One-Hot-Encoding on categorical numbers.**

```python
from sklearn.preprocessing import OneHotEncoder
hotencoder = OneHotEncoder()
encoded = hotencoder.fit_transform(df_dummies.RegionId.values.reshape(-1,1)).toarray()
encoded
```

```python
encoded.shape
```

Let us now add this back to df_dummies dataframe.

```python
df_encoded = pd.DataFrame(encoded, columns = ["RegionId_"+str(int(i)) for i in range(encoded.shape[1])])
```

```python
df_encoded.head()
```

```python
df_encoded.shape
```

```python
df_dummies = pd.concat([df_dummies, df_encoded], axis=1)
```

```python
df_dummies.head(5)
```


### Normalization and Scaling


In this method, we convert variables with different scales of measurements into a single scale. 

**StandardScaler** normalizes the data using the formula (x-mean)/standard deviation

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scale = StandardScaler()
```

```python
std_scale
```

```python
data1['Bedroom_Stdscale'] = std_scale.fit_transform(data1[['Bedroom']])
```

```python
data1['Bedroom_Stdscale'].head(5)
```

```python
data1['Bedroom_Stdscale'].min(), data1['Bedroom_Stdscale'].max()
```

```python
data1['Bedroom_Stdscale'].describe()
```

**MinMaxScaler** normalizes the data using the formula (x - min)/(max - min)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax_scale = MinMaxScaler()
```

```python
minmax_scale
```

```python
data1['Car_MinMaxScale'] = minmax_scale.fit_transform(data1[['Car']])
```

```python
data1['Car_MinMaxScale'].head(5)
```

```python
data1['Car_MinMaxScale'].min(), data1['Car_MinMaxScale'].max()
```

```python
data1['Car_MinMaxScale'].describe()
```

**Log Transformation** 

We are going to write a custom transformer to perform log transformation

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
log_transformer = FunctionTransformer(np.log1p)
```

```python
log_transformer
```

```python
data1['Distance'].isnull().sum()
```

```python
data1['Distance']= data1['Distance'].fillna(data1['Distance'].mean())
```

```python
data1['Distance_logtransform'] = log_transformer.fit_transform(data1[['Distance']])
```

```python
data1['Distance_logtransform'].head(5)
```

```python
data1['Distance_logtransform'].describe()
```

**Exponential Transformation** 

```python
exp_transformer = FunctionTransformer(np.exp)
```

```python
exp_transformer
```

```python
data1['Rooms_exptransform'] = exp_transformer.fit_transform(data1[['Rooms']])
```

```python
data1['Rooms_exptransform'].head(5)
```
